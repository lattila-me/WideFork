## Histogram Analyzis

```python
# Histogram analyzis function
import matplotlib.pyplot as plt
import scipy.stats as scipy

def plotHist(data, title="", bins=8):
    fig, ax = plt.subplots()
    ax.hist(
        data[title],        
        bins=bins,
        linewidth=0.5,
        facecolor="green",
        edgecolor="white",
        alpha=0.75        
    )   

    plt.title(title)
    plt.xlabel(title)
    plt.ylabel("Frequency")
    plt.grid(True)    

    plt.savefig(f"content/data_analyzis/Hist_{title}.pdf")   
```

```python
plotHist(data, title="OSSZEG", bins=30)
```

## Handling Outliers

```python
# Handling outliers using Z-score -> This function will maximize the value of datapoints to a pre-defined max Z-score value (cutoff -> defaults to 3)
import scipy.stats as stats

def reshape(data, columns, cutoff=3):
    for col in columns:        
        data["zscores"] = stats.zscore(data[col])
        x_max = (cutoff + data[col].mean()) * data[col].std()        
        data["factor"] = data[col] / x_max        
        data[col] = data.apply(lambda x: x[col] if x["zscores"] <= cutoff else x[col] / x["factor"], axis="columns")
        data.drop(["zscores"], inplace=True, axis="columns")
        data.drop(["factor"], inplace=True, axis="columns")
        print(f"Outlier cutoff absolute limit (x_max) for column [{col}]: {x_max}")  
```

```python
columns_outliers = [
    'OSSZEG'    
]
```

```python
reshape(data, columns_outliers, 3)
```
